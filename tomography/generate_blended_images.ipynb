```python
num_slices = 24
num_blends = 6 *3
num_stills = 5 *3

blend_fractions = np.linspace(0.0, 1.0, num_blends)

import tomo_data
reload(tomo_data)
```

```python
voorkant_images = [tomo_data.normed_fn_format % ('v', x) for x in range(1, num_slices + 1)]
achterkant_images = [tomo_data.normed_fn_format % ('a', x) for x in range(1, num_slices + 1)]

cur_kant = 'snel_a'
cur_images = achterkant_images
```

```python
image_pairs = list(zip(cur_images[:-1], cur_images[1:]))
```

```python
out_i = 1

for (imgf1, imgf2) in image_pairs:
    pil_img1 = Image.open(imgf1)
    pil_img2 = Image.open(imgf2)
    
    for still_i in range(num_stills):
        pil_img1.save(tomo_data.blended_fn_format % (cur_kant, out_i))
        
        print("Saved image %d (still)" % out_i)
        out_i += 1
    
    for bf in blend_fractions:
        blended_img = Image.blend(pil_img1, pil_img2, bf)
        blended_img.save(tomo_data.blended_fn_format % (cur_kant, out_i))
        
        print("Saved image %d (blend)" % out_i)
        out_i += 1
        
for still_i in range(num_stills):
    pil_img2.save(tomo_data.blended_fn_format % (cur_kant, out_i))

    print("Saved image %d (still)" % out_i)
    out_i += 1
```
